```python
import yfinance as yf
import pandas as pd
import seaborn as sns
```

# 1st task

Use API of yahoo finance to download Apple's ticker

```python
hist = yf.download('aapl', period = '1y')
hist
```

# 2nd task

Calculate the coefficient and multiply it by each column

```python
coeff = hist['Adj Close'] / hist['Close']
hist = hist.drop('Adj Close', axis = 1)
```

```python
adjhist = pd.DataFrame(list(map(lambda x: hist[x]*coeff, hist.columns[:4]))).transpose().rename(columns = {0:'Open',1:'High',2:'Low',3:'Close'})
adjhist['Volume'] = hist['Volume']
```

```python
adjhist
```

# 3rd task

Construct line and histogram plots for the features

```python
sns.lineplot(adjhist[['Open','High', 'Low', 'Close']])
```

```python
sns.histplot(data = adjhist['Volume'])
```

# 4th task

Use the API again to download QQQ ticker

```python
qqq = yf.download('qqq', period = '1y')
qqq
```

Now check 3 conditions
1. QQQ and AAPL trade at the same dates
2. Low <= Open <= High
3. Low <= Close <= High

```python
all(adjhist.index == qqq.index)
```

```python
all(adjhist['Low'] <= adjhist['Open']) and all(adjhist['Open'] <= adjhist['High'])
```

```python
all(adjhist['Low'] <= adjhist['Close']) and all(adjhist['Close'] <= adjhist['High'])
```

```python
set(adjhist.index) & set(qqq.index) == set(qqq.index)
```

The data is correct

# 5th task

Make 3 Simple Moving Average and plot them

```python
adjhist['SMA5'] = adjhist['Close'].rolling(5).mean()
adjhist['SMA10'] = adjhist['Close'].rolling(10).mean()
adjhist['SMA20'] = adjhist['Close'].rolling(20).mean()
```

```python
sns.lineplot(adjhist[['Close', 'SMA5', 'SMA10', 'SMA20']])
```
